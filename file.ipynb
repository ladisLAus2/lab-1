```python
import numpy as np
import csv
import matplotlib.pyplot as plt
with open("./lab_1_train.csv",'r') as train_set:
  csvreader = csv.reader(train_set)
  data_train = [row for row in csvreader]
data_train.pop(0)
x_set = [float(row[1]) for row in data_train]
y_set = [float(row[2]) for row in data_train]
```

```python
plt.title("train_dataset")
plt.scatter(x_set, y_set,c='green');
plt.show()
```

```python
def linear_Regression(b0, b1, x):
  return b0 + b1 * x;

def loss_function(b0,b1,x, y):
  result = 0
  for i in range(len(x)):
    result += (y[i] - (b0 + b1 * x[i])) ** 2
  return result / len(x)
```

```python

tolerance = 0.000001
L = 0.003
b0 = float(abs(np.random.randn(1)))
b1 = float(abs(np.random.randn(1)))
loss_value = loss_function(b0,b1,x_set,y_set)
delta = loss_value
epoch = 0
while delta > tolerance:
  y_predicted = [linear_Regression(b0,b1, x) for x in x_set]
  
  y_predicted_diference = [(real - predicted) for (real, predicted) in zip(y_set, y_predicted)]

  der_b0 = -2 * np.sum(y_predicted_diference) / len(y_set)

  der_b1 = - (2 / len(y_predicted_diference)) * np.sum([(x*y) for (x,y) in zip(x_set, y_predicted_diference) ])

  b0 = b0 - L * der_b0
  b1 = b1 - L * der_b1

  print(str(epoch) + " - b0 = " + str(b0) + "  b1 = " + str(b1) + "  lossFunction = " + str(loss_value))
  delta = loss_value - loss_function(b0,b1, x_set, y_set)
  loss_value = loss_function(b0, b1, x_set, y_set)
  epoch+=1
```

```python
with open("./lab_1_test.csv",'r') as test_set:
  csvreader = csv.reader(test_set)
  data_test = [row for row in csvreader]
data_test.pop(0)
x_set_test = [float(row[1]) for row in data_test]
y_set_test = [float(row[2]) for row in data_test]
plt.title("test_dataset")
plt.scatter(x_set_test, y_set_test,c='blue');
plt.show()
```
